```python
import numpy as np, pandas as pd
df = pd.DataFrame({'A': [np.random.randint(1,100) for i in range(4)],
                   'B': pd.date_range(start='20130101', periods=4, freq='D'),
                   'C': pd.Series([1, 2, 3, 4],index=['zhang', 'li', 'zhou', 'wang'],dtype='float32'),
                   'D': np.array([3] * 4,dtype='int32'),
                   'E': pd.Categorical(["test","train","test","train"]),
                   'F': 'foo'})
df
```

```python
df[df.A>50]
```

```python
df[df['A'].isin(range(20,40))]
```

```python
df.nlargest(3, ['A'])
```

```python
df1 = df.reindex(columns=list(df.columns) + ['G'])  # 缺失值都默认为NAN
df1
```

```python
df1.iat[0, 6]  # at指具体位置 包含i的代表int 用数字代表索引
```

```python
df1.iloc[:,6]  # loc指一行一列 或者也可指具体位置
```

```python
df1.at['zhang', 'G']  # 不包含i的代表要用索引访问
```

```python
df1.at[0, 6]
```

```python
df1.iat[0, 6]=3
df1
```

```python
df1.dropna()         # 返回不包含缺失值的行
```

```python
df1['G'].fillna(5, inplace=True)    # 使用指定值填充缺失值
df1
```

```python
data = pd.DataFrame({'k1': ['one'] * 3 + ['two'] * 4,
                     'k2': [1, 1, 2, 3, 3, 4, 4]})
data
```

```python
data.duplicated()      # 检查重复行
```

```python
data.drop_duplicates()        # 返回新数组，删除重复行
```

```python
data  # 注意之前的操作并没有在data上改变 而是构造了一个新数组
```

```python
data.drop_duplicates(['k1'])  # 删除k1列的重复数据
```

```python
data.drop_duplicates(['k1'], keep='last')  # 保留的是最后一个而不是第一个出现的
```
